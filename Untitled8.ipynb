```python
import numpy as np
import matplotlib.pyplot as plt

Train_cluster_X1 = []
Train_cluster_Y1 = []
Train_cluster_X2 = []
Train_cluster_Y2 = []

c1 = (0.25,0.75)
c2 = (0.75,0.25)

nc1 = 0
nc2 = 0
np.random.seed(1)

while True:
    t = np.random.rand(2)
    if (pow(t[0]-c1[0],2) + pow(t[1]-c1[1],2) < 0.04):
        Train_cluster_X1.append(list(t))
        Train_cluster_Y1.append(1)
        nc1+=1
    if (pow(t[0]-c2[0],2) + pow(t[1]-c2[1],2) < 0.04):
        Train_cluster_X2.append(list(t))
        Train_cluster_Y2.append(0)
        nc2 += 1

    if nc1>+nc2>=400:
        break

Train_cluster_X1 = np.asarray(Train_cluster_X1)
Train_cluster_X2 = np.asarray(Train_cluster_X2)

plt.axis([0,1,0,1])
plt.scatter(Train_cluster_X1[:,0],Train_cluster_X1[:,1],color='b',marker='o')
plt.scatter(Train_cluster_X2[:,0],Train_cluster_X2[:,1],color='r',marker='o')
plt.show()
```

```python
Test_cluster_X1 = []
Test_cluster_Y1 = []
Test_cluster_X2 = []
Test_cluster_Y2 = []

nc1 = 0
nc2 = 0
np.random.seed(5)

c1 = (0.15,0.65)
c2 = (0.85,0.35)

while True:
    t = np.random.rand(2)
    if (pow(t[0]-c2[0],2) + pow(t[1]-c2[1],2) < 0.09):
        Test_cluster_X2.append(list(t))
        Test_cluster_Y2.append(0)
        nc2 += 1
    if (pow(t[0]-c1[0],2) + pow(t[1]-c1[1],2) < 0.09):
        Test_cluster_X1.append(list(t))
        Test_cluster_Y1.append(1)
        nc1+=1

    if nc1+nc2 >=400:
        break

Test_cluster_X1 = np.asarray(Test_cluster_X1)
Test_cluster_X2 = np.asarray(Test_cluster_X2)

plt.scatter(Train_cluster_X1[:,0],Train_cluster_X1[:,1],color='b',marker='o')
plt.scatter(Train_cluster_X2[:,0],Train_cluster_X2[:,1],color='r',marker='o')
plt.scatter(Test_cluster_X1[:,0],Test_cluster_X1[:,1],color='g',marker='o')
plt.scatter(Test_cluster_X2[:,0],Test_cluster_X2[:,1],color='y',marker='o')
plt.show()
```

```python
Train_X =[]
Train_Y =[]
Test_X = []
Test_Y = []
Train_X.extend(Train_cluster_X1.tolist())
Train_X.extend(Train_cluster_X2.tolist())
Train_Y.extend(Train_cluster_Y1)
Train_Y.extend(Train_cluster_Y2)
Test_X.extend(Test_cluster_X1.tolist())
Test_X.extend(Test_cluster_X2.tolist())
Test_Y.extend(Test_cluster_Y1)
Test_Y.extend(Test_cluster_Y2)
```

```python
def sigmoid(w,b,tx):
    z = w[0]*tx[0] + w[1]*tx[1] +b
    return 1/(1+np.exp(-z))

def lossfunc(py,y):
    sum = 0
    for i in range(len(py)):
        sum += -y[i]*np.log(py[i])-(1-y[i])*np.log(1-py[i])
    return sum/len(py)

def gradient_decent(w,b,rate):
    sum = 0
    for i in range(len(Train_X)):
        sum+= -Train_Y[i]*(sigmoid(w,b,Train_X[i])*(1-sigmoid(w,b,Train_X[i])))*Train_X[i][0]/sigmoid(w,b,Train_X[i])-(1-Train_Y[i])*(-sigmoid(w,b,Train_X[i])*(1-sigmoid(w,b,Train_X[i])))*Train_X[i][0]/(1-sigmoid(w,b,Train_X[i]))
    rw1 = w[0] - rate*(sum/len(Train_X))

    sum = 0
    for i in range(len(Train_X)):
        sum += -Train_Y[i] * (sigmoid(w, b, Train_X[i]) * (1 - sigmoid(w, b, Train_X[i]))) * Train_X[i][1] / sigmoid(w,b,Train_X[i]) - (1 - Train_Y[i]) * (-sigmoid(w, b, Train_X[i]) * (1 - sigmoid(w, b, Train_X[i]))) * Train_X[i][1] / (1 - sigmoid(w, b, Train_X[i]))
    rw2 = w[1] - rate*(sum/len(Train_X))

    sum = 0
    for i in range(len(Train_X)):
        sum += -Train_Y[i] * (sigmoid(w, b, Train_X[i]) * (1 - sigmoid(w, b, Train_X[i])))/ sigmoid(w,b,Train_X[i]) - (1 - Train_Y[i]) * (-sigmoid(w, b, Train_X[i]) * (1 - sigmoid(w, b, Train_X[i])))/ (1 - sigmoid(w, b, Train_X[i]))
    rb = b - rate*(sum/len(Train_X))
    return (rw1,rw2,rb)
```

```python
w = [0.1,0.9]
b = 0.1
learning_rate = 0.2
epoches = 70

train_py = []
train_pred = []
test_py = []
test_pred = []

listOfTrainLoss =[]
listOfTrainAcc = []
listOfTestLoss = []
listOfTestAcc = []
for i in range(epoches):
    new = gradient_decent(w,b,learning_rate)
    w[0] = new[0]
    w[1] = new[1]
    b = new[2]

    train_py.clear()
    test_py.clear()
    for j in range(len(Train_X)):
        train_py.append(sigmoid(w,b,Train_X[j]))
    j=0
    for j in range(len(Test_X)):
        test_py.append(sigmoid(w,b,Test_X[j]))

    train_pred.clear()
    test_pred.clear()
    for k in range(len(Train_Y)):
        if train_py[k] > 0.5 :
            train_pred.append(1)
        else :
            train_pred.append(0)
    k = 0
    for k in range(len(Test_Y)):
        if test_py[k] > 0.5:
            test_pred.append(1)
        else:
            test_pred.append(0)


    train_count = 0
    test_count = 0
    for l in range(len(Train_Y)):
        if train_pred[l] == Train_Y[l]:
            train_count+=1
    l=0
    for l in range(len(Test_Y)):
        if test_pred[l] == Test_Y[l]:
            test_count+=1

    train_acc = train_count/len(Train_Y)
    train_loss = lossfunc(train_py,Train_Y)
    test_acc = test_count / len(Test_Y)
    test_loss = lossfunc(test_py,Test_Y)

    listOfTrainAcc.append([i,train_acc])
    listOfTrainLoss.append([i,train_loss])
    listOfTestAcc.append([i,test_acc])
    listOfTestLoss.append([i,test_loss])

    print("epoch %4d --- training loss : %f , training acc : %f , test loss : %f , test acc : %f"%(i,train_loss,train_acc,test_loss,test_acc))
    #print(train_pred)
```

```python
listOfTrainLoss = np.asarray(listOfTrainLoss)
listOfTrainAcc = np.asarray(listOfTrainAcc)
listOfTestLoss = np.asarray(listOfTestLoss)
listOfTestAcc = np.asarray(listOfTestAcc)
plt.plot(listOfTrainLoss[:,0],listOfTrainLoss[:,1],c="y",label='TrainLoss')
plt.plot(listOfTestLoss[:,0],listOfTestLoss[:,1])
plt.plot(listOfTrainAcc[:,0],listOfTrainAcc[:,1],c="r")
plt.plot(listOfTestAcc[:,0],listOfTestAcc[:,1],c="b")
plt.show()
```
