```python
import numpy as np
import matplotlib.pyplot as plt

#-------------- Generate Training data ------------------------

Train_cluster_X1 = []
Train_cluster_Y1 = []
Train_cluster_X2 = []
Train_cluster_Y2 = []

c1 = (0.25,0.75)
c2 = (0.75,0.25)

nc1 = 0
nc2 = 0
np.random.seed(1)

while True:    #---Create 400 Training Data (upper left cluster, lower right cluster)
    t = np.random.rand(2)
    if (pow(t[0]-c1[0],2) + pow(t[1]-c1[1],2) < 0.04):
        Train_cluster_X1.append(list(t))
        Train_cluster_Y1.append(1)
        nc1+=1
    if (pow(t[0]-c2[0],2) + pow(t[1]-c2[1],2) < 0.04):
        Train_cluster_X2.append(list(t))
        Train_cluster_Y2.append(0)
        nc2 += 1

    if nc1>+nc2>=400:
        break

Train_cluster_X1 = np.asarray(Train_cluster_X1)
Train_cluster_X2 = np.asarray(Train_cluster_X2)

plt.axis([0,1,0,1])
plt.scatter(Train_cluster_X1[:,0],Train_cluster_X1[:,1],color='b',marker='o',label='TrainData1')
plt.scatter(Train_cluster_X2[:,0],Train_cluster_X2[:,1],color='r',marker='o',label='TrainData0')
plt.legend(loc="lower left")
plt.show()
```

```python
#-------------- Generate Testing data --------------------------

Test_cluster_X1 = []
Test_cluster_Y1 = []
Test_cluster_X2 = []
Test_cluster_Y2 = []

nc1 = 0
nc2 = 0
np.random.seed(5)

c1 = (0.15,0.65)
c2 = (0.85,0.35)

while True: #---Create 400 Test Data (upper left cluster, lower right cluster)
    t = np.random.rand(2)
    if (pow(t[0]-c2[0],2) + pow(t[1]-c2[1],2) < 0.09):
        Test_cluster_X2.append(list(t))
        Test_cluster_Y2.append(0)
        nc2 += 1
    if (pow(t[0]-c1[0],2) + pow(t[1]-c1[1],2) < 0.09):
        Test_cluster_X1.append(list(t))
        Test_cluster_Y1.append(1)
        nc1+=1

    if nc1+nc2 >=400:
        break

Test_cluster_X1 = np.asarray(Test_cluster_X1)
Test_cluster_X2 = np.asarray(Test_cluster_X2)

plt.scatter(Train_cluster_X1[:,0],Train_cluster_X1[:,1],color='b',marker='o',label='TrainData1')
plt.scatter(Train_cluster_X2[:,0],Train_cluster_X2[:,1],color='r',marker='o',label='TrainData0')
plt.scatter(Test_cluster_X1[:,0],Test_cluster_X1[:,1],color='g',marker='o',label='TestData1')
plt.scatter(Test_cluster_X2[:,0],Test_cluster_X2[:,1],color='y',marker='o',label='TestData0')
plt.legend(loc="lower left")
plt.show()
```
